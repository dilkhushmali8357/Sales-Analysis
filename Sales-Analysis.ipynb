```python
# import python libraries

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("All libraries imported successfully!")
```

```python
# import csv file
df = pd.read_csv('Sales-Data.csv', encoding= 'unicode_escape')
```

```python
#drop unrelated/blank columns
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
df.shape
```

```python
#check for null values
pd.isnull(df).sum()
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
# change data type
df['Amount'] = df['Amount'].astype('int')
```

```python
#Exploratory Data Analysis
```

```python
##Gender
```

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen,hue= 'Gender')
```

```python
### Insight

The analysis shows that female customers constitute the majority of buyers. Additionally, the total sales amount generated by female customers is significantly higher than that of male customers, indicating stronger purchasing power among female buyers.
```

```python
##Age
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group

sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Age Group',y= 'Amount',data = sales_age,hue='Age Group')
```

```python
### Insight

The analysis indicates that female customers aged 26–35 years constitute the largest group of buyers, suggesting that this demographic contributes significantly to overall sales.
```

```python
##State
```

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders',hue='State')
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='State',y='Amount',hue='State')
```

```python
### Insight

The analysis shows that Uttar Pradesh contributes the highest number of orders and total sales amount, followed by Maharashtra and Karnataka. These states are the key contributors to overall business revenue.
```

```python
## Marital Status
```

```python
ax = sns.countplot(data = df, x = 'Marital_Status',hue= 'Marital_Status')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Marital Status

sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

```python
### Insight

The analysis shows that married women represent the largest customer segment and exhibit the highest purchasing power, contributing significantly to overall sales.
```

```python
## Occupation
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation',hue='Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Occupation

sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount',hue= 'Occupation')
```

```python
### Insight

The analysis indicates that customers working in the IT, Healthcare, and Aviation sectors constitute the largest share of buyers, suggesting that these professions are key contributors to overall sales.
```

```python
## Product Category
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category',hue='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Product Category

sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount',hue='Product_Category')
```

```python
### Insight

The analysis indicates that Food, Clothing, and Electronics are the most frequently purchased product categories, making them the primary contributors to overall sales volume.
```

```python
# top 10 most sold products

sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders',hue='Product_ID')
```

```python
# top 10 most sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

```python
## Conclusion:

### Married women aged 26–35 years from Uttar Pradesh, Maharashtra, and Karnataka, particularly those working in the IT, Healthcare, and Aviation sectors, are the most likely to purchase products. Their purchases are primarily concentrated in the Food, Clothing, and Electronics categories, making them the key customer segment.
```
